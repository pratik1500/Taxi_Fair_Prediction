```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
'''Fetching cleaned dataset'''
df = pd.read_csv('D:\Projects\coditas_project\dataset\data_cleaned.csv')
df.head()
```

```python
'''Droping the unuseful columns'''
drop_columns=['key','pickup_datetime','pickup_date', 'index']
df_after_drop = df.drop(drop_columns,axis=1)
```

```python
df_after_drop.head()
```

```python
x=df_after_drop.drop(['fare_amount'],axis=1) # Selecting all non-target variables
y = df_after_drop['fare_amount'] # Target Variable
```

```python
''''split data into train and validation data'''
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2) # Spliting data into train and test
print("Number of records in training data ",X_train.shape[0])
print("Number of records in validation data ",X_test.shape[0])
```

Analysing the result using rmse 
Which is root mean square error

```python
'''Training a simple linear regression model'''
lm = LinearRegression() # Defining linear regression model
lm.fit(X_train,y_train) # Fit on the training dataset
y_pred=lm.predict(X_test) # Predicting test data 
lm_rmse=np.sqrt(mean_squared_error(y_pred, y_test)) 
print("RMSE for Linear Regression is ",lm_rmse)
```

Random Forest

```python
'''Traing the Radom Forest Regressor'''
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(max_depth=140,max_features='auto',min_samples_leaf=10,n_estimators=400) # Defining Regression model after hyperparameter tunning
clf.fit(X_train, y_train) # Training the model
y_pred = clf.predict(X_test) # Predicting on the prediction dataset 
rf_rmse = np.sqrt(mean_squared_error(y_pred, y_test)) # Calculating rmse ===> Root mean squared error
print("RMSE FOR RANDOM FOREST IS",rf_rmse)
```
